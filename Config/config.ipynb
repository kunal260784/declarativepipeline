```python
dbutils.widgets.text("env", "dev")
env = dbutils.widgets.get("env")

dbutils.widgets.text("catelog","myproject")
catelog = dbutils.widgets.get("catelog")

dbutils.widgets.text("schema","etlconfig")
schm = dbutils.widgets.get("schema")
```

```python

ctg = f""" 
        CREATE CATALOG IF NOT EXISTS {catelog}_{env}
        """
spark.sql(ctg)

use = f""" USE CATALOG {catelog}_{env}
        """

spark.sql(use)

schm_qry = f""" CREATE SCHEMA IF NOT EXISTS {catelog}_{env}.{schm}
        """

spark.sql(schm_qry)

qry = f""" CREATE TABLE IF NOT EXISTS {catelog}_{env}.{schm}.source
(
  id string,
  name string,
  shortname string,
  description string,
  createddt timestamp,
  updateddt timestamp,
  createdby string,
  updatedby string
)
"""

spark.sql(qry)


qry = f""" CREATE OR REPLACE TABLE {catelog}_{env}.{schm}.sourcetables
(
  id string,
  sourceid string,
  tablename string,
  description string,
  catelog string,
  schema string,
  filename string,
  filepath string,
  format string,
  partition string,
  createddt timestamp,
  updateddt timestamp,
  createdby string,
  updatedby string
)
"""

spark.sql(qry)
```
